## Using the List 'raw_info', created a Nested List where the information of each student is stored in sub-lists
* I observed that the below sequence of Eight Attributes repeat in the raw_info List:
 * Class
 * Section
 * Rollno
 * Studentname
 * Gender
 * English
 * Mathematics
 * Science

```python
count = 0
nested_list = []
my_list = []
for i in raw_info:
    my_list.append(i)
    count = count + 1
    if(count%8 == 0):
        nested_list.append(my_list)
        count = 0
        my_list = []

del nested_list[0]                   #deleted the 1st entry of the list as it contain headers, I donr this for better implementation of data in next steps.

nested_list

```

## Used the created Nested List to correct the format of Numbers shown as Strings

```python
nested_list_1 = nested_list
nested_list_2 = []

for i in nested_list_1:
    my_list = []
    for j in i:
        if (j.isdigit() == 1):
            my_list.append(int(j))
        else:
            my_list.append(j)
    nested_list_2.append(my_list)

nested_list = nested_list_2

nested_list




```

### Using the updated Nested List, stored the information in a Nested Dictionary which follows the structure --> {Class : {Section : { Student Name : [Subject Marks] } } }

* Here I have declared 2 lists which contains class and section by observing the data in the raw_info to create a dictionary in above format.



```python
dict_class= {}
dict_section = {}
my_dict = {}

classs = [8,9,10]
section = ['A','B']

for i in classs:
    dict_section = {}
    for j in section:
        my_dict = {}
        for k in nested_list:
            if(k[0]== i and k[1] == j):
                my_dict.update({k[3]:k[5:8]})
            dict_section.update({j:my_dict})
    dict_class.update({i:dict_section})


dict_class












```

### Created a Function which takes Class (C), Section (S) and Student Name (N) as input and prints subject wise marks of that student.

```python
def show_sub_scores(C,S,N):
    print(f'Name:{N}, Class:{C}-{S}')
    print(f'English : {dict_class[C][S][N][0]}')
    print(f'Mathematics : {dict_class[C][S][N][1]}')
    print(f'Science : {dict_class[C][S][N][2]}')
```

### Made changes in the Nested Dictionary such that the individual subject scores (a list) is replaced with the Total scores (sum of these three subjects)

* Here I have used the code from the previous cell as only one change was full filling my purpose.


```python
dict_class= {}
dict_section = {}
my_dict = {}

classs = [8,9,10]
section = ['A','B']

for i in classs:
    dict_section = {}
    for j in section:
        my_dict = {}
        for k in nested_list:
            if(k[0]== i and k[1] == j):
                my_dict.update({k[3]:k[5]+k[6]+k[7]})             #Change#
            dict_section.update({j:my_dict})
    dict_class.update({i:dict_section})


dict_class
```

## Found the Highest values of these Total Scores for each Class and stored them in a Dictionary
* For the Dictionary , the 'Key' and 'Value' parts are Class and Max of Total Scores respectively


```python
max_scores = {}
classs = [8,9,10]
section = ['A','B']

for i in classs:
    my_list = []
    for j in section:
        a=(max(dict_class[i][j].values()))
        my_list.append(a)
    max_scores.update({i:max(my_list)})

max_scores

```

## Found Top Scorers for each Class and stored them in a Dictionary (taken care of multiple top scorers)
* For the Dictionary the 'Key' part is the Class
* The 'Value' part is a List, which stores the name of Top Scorer(s)


```python
class_toppers = {}




classs = [8,9,10]

section = ['A','B']

for i in classs:
    my_list = []
    for j in section:
        my_dict = dict_class[i][j]
        for k,l in my_dict.items():
            if (l == max_scores[i]):
                my_list.append(k)
    class_toppers.update({i:my_list})

class_toppers




```
